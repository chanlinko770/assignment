```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(1)
names = [f"Student {i}" for i in range(1, 21)]
math = np.random.randint(50, 100, size=20)
english = np.random.randint(40, 90, size=20)
science = np.random.randint(45, 95, size=20)
attendance = np.random.randint(60, 100, size=20)
df = pd.DataFrame({
    'Name': names,
    'Math_Score': math,
    'English_Score': english,
    'Science_Score': science,
    'Attendance_%': attendance
})

df['Average_Score'] = df[['Math_Score', 'English_Score', 'Science_Score']].mean(axis=1).round(2)
```

```python
print("***Student Performance Data:")
print(df.head())
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(df['Name'], df['Math_Score'], marker='o', label='Math')
plt.plot(df['Name'], df['Science_Score'], marker='s', label='Science')
plt.title('Math & Science Scores')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.bar(df['Name'], df['English_Score'], color='skyblue')
plt.title('English Scores')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x='Math_Score', y='Science_Score', data=df)
plt.title('Math vs Science Score')
plt.grid(True)
plt.show()
```

```python
top5 = df.sort_values('Average_Score', ascending=False).head(5)
sns.barplot(x='Name', y='Average_Score', data=top5, palette='mako')
plt.title('Top 5 Students by Average Score')
plt.xticks(rotation=30)
plt.show()
```

```python
sns.boxplot(data=df[['Math_Score', 'English_Score', 'Science_Score']])
plt.title('Score Distribution')
plt.show()
```

```python
sns.heatmap(df[['Math_Score', 'English_Score', 'Science_Score', 'Attendance_%']].corr(), annot=True, cmap='coolwarm')
plt.title('Score Correlation Heatmap')
plt.show()
```
